```python
def string_to_time(string):
    try:
        try:
            return datetime.strptime(string.strip(),'%B %d, %Y')
        except:
            return datetime.strptime(string.strip(),'%b %d, %Y')
    except:
        try:
            return datetime.strptime(string.strip(),'%d-%b-%y')
        except:
            return string

news['date'] = news['date'].apply(string_to_time)
```
